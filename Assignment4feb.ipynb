```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = {
    'Singer': [
        'A.R. Rahman', 'Neha Kakkar', 'Atif Aslam', 'Shilpa Rao', 'Udit Narayan', 
        'Arijit Singh', 'Shreya Ghoshal', 'Sonu Nigam', 'Lata Mangeshkar', 'Kishore Kumar', 
        'Kunal Ganjawala', 'Alka Yagnik', 'Himesh Reshammiya', 'Rahat Fateh Ali Khan', 'Mohammad Rafi',
        'Sukhwinder Singh', 'Richa Sharma', 'Javed Ali', 'Sunidhi Chauhan', 'Ankit Tiwari'
    ],
    'Years_Active': [
        25, 12, 18, 15, 35, 
        15, 20, 30, 70, 40, 
        20, 30, 15, 25, 60, 
        25, 20, 18, 15, 10
    ],
    'Number_of_Songs': [
        1000, 400, 550, 300, 1200, 
        500, 700, 600, 1000, 800, 
        650, 800, 400, 550, 1000, 
        650, 500, 350, 300, 250
    ],
    'Awards_Won': [
        70, 30, 45, 25, 90, 
        50, 60, 40, 100, 80, 
        60, 55, 30, 70, 75, 
        45, 50, 35, 20, 15
    ]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Plot the bar chart
plt.figure(figsize=(10,6))  # Set the figure size
plt.bar(df['Singer'], df['Number_of_Songs'], color='lightgreen')

# Rotate the x-axis labels to avoid overlapping
plt.xticks(rotation=90)

# Add title and labels
plt.title('Number of Songs by Singer')
plt.xlabel('Singer')
plt.ylabel('Number of Songs')

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels

plt.show()
```
